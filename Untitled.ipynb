```python
# Graph implementation
class Graph:
    graph = {}
    def add_edges(self,node,another_node):
        if node not in self.graph:
            self.graph[node]=[another_node]
        else:
            self.graph[node].append(another_node)
    def showgraph(self):
        for node in self.graph:
            for another_node in self.graph[node]:
                print(node,another_node)
    def find_path(self,start,end,path=[]):
        path = path + [start]    
        if start==end:
            return path
        for node in self.graph[start]:
            if node not in path:
                newPath=self.find_path(node,end,path)
                if newPath:
                    return newPath
                return None             
```

```python
#Sorting 
# Bubble sort 
# Put the largest value at the end 
# Time Comlexity of bubble sort is O(n^2) == O((n-1)(n-2))
def Bubble_sort(A):
    # This loop  for passes
    for i in range(len(A)-1):
        swapped = False
        for j in range(len(A)-1-i):
            if A[j] > A[j+1]:
                swapped =True
                A[j], A[j+1] = A[j+1],A[j]
        if not swapped:
            break
    return A


# Insertion Sort
def insertion_sort(A):
    for i in range(1,len(A)):
        j =i
        while (j>0 and A[j-1])> A[j]:
            A[j],A[j-1] = A[j-1],A[j]
            
            j-=1
    return A


# Truely didn't understand
#Bucket Sort
SIZE_OF_BUCKET = 6
def bucket_sort(A):
    bucket_count = (max(A)-min(A))//SIZE_OF_BUCKET +1
    buckets = [[] for _ in range(bucket_count)]
    for i in range(len(A)):
        buckets[(A[i]-min(A))//SIZE_OF_BUCKET].append(A[i])
    
    return sorted(
    list(buckets[i][j] for i in range(len(buckets)) for j in range(len(buckets[i]))))

# merge sort
# it works on left and right side sorting 
# follow recursion to solve left and right part 
def Merge_sort(A):
    
    def merge(left,right):
        result = []
        while left and right:
            if left[0]<= right[0]:
                result.append(left.pop(0))
            else:
                result.append(right.pop(0))       
        return result+left+right
    if len(A) <=1:
        return A
    mid = len(A)//2
    return merge(Merge_sort(A[:mid]),Merge_sort(A[mid:]))

# Quick Sort
# last element use as pivot 
def quick_sort(A):
    if len(A)<=1:
        return A
    pivot = A.pop() # last element is pivot
    great, less = [],[]
    for element in A:
        if element < pivot:
            less.append(element)
        else:
            great.append(element)
        
    return quick_sort(less) + [pivot]+ quick_sort(great)    
```

```python
#DFS,BFS,Dijkdtra, prims,kruskal, bellman-ford,floyd-warshall

# BFS
G = {
    "A": ["B", "C", "E"],
    "B": ["A", "D", "E"],
    "C": ["A", "F", "G"],
    "D": ["B"],
    "E": ["A", "B", "D"],
    "F": ["C"],
    "G": ["C"],
}
def bfs(G,start):
    visited = set()
    queue = [start]
    
    visited.add(start)
    while queue:
        node = queue.pop(0)
        for w in G[node]:
            if w not in visited:
                queue.append(w)
                visited.add(w)
    return visited              
                
def bfs_shortest_path(G,start,end):
    visited = []
    queue = [[start]]
    
    if start==end:
        return
    
    while queue:
        path = queue.pop(0)
        node = path[-1]
        if node not in visited:
            for v in G[node]:
                newpath = list(path)
                newpath.append(v)
                queue.append(newpath)
                if v == end:
                    return newpath
                
            visited.append(node)
            
            
    return "There is no connecting path"

def bfs_shortest_distance(G,start,end):
    visited = [start]
    queue = [start]
    
    counter= {start:0,end:0}
    while queue:
        node = queue.pop(0)
        '''if node == target:
            dist[target] = (
                dist[node] if dist[target] == -1 else min(dist[target], dist[node])
            )'''
        for v in G[node]:
            if v not in visited:
                visited.append(v)
                queue.append(v)
                counter[v] = counter[node] +1
   
    return counter[end]
        
    
```

```python
#DFS
def dfs(G,start):
    visited = set()
    stack = [start]
    
    while stack:
        node = stack.pop()
        visited.add(node)
        for v in G[node]:
            if v not in visited:
                stack.append(v)
   
    return visited  
    
```

```python
class BinaryHeap:
    def __init__(self):
        self.items = []
 
    def size(self):
        return len(self.items)
 
    def parent(self, i):
        return (i - 1)//2
 
    def left(self, i):
        return 2*i + 1
 
    def right(self, i):
        return 2*i + 2
 
    def get(self, i):
        return self.items[i]
 
    def get_max(self):
        if self.size() == 0:
            return None
        return self.items[0]
 
    def extract_max(self):
        if self.size() == 0:
            return None
        largest = self.get_max()
        self.items[0] = self.items[-1]
        del self.items[-1]
        self.max_heapify(0)
        return largest
 
    def max_heapify(self, i):
        l = self.left(i)
        r = self.right(i)
        if (l <= self.size() - 1 and self.get(l) > self.get(i)):
            largest = l
        else:
            largest = i
        if (r <= self.size() - 1 and self.get(r) > self.get(largest)):
            largest = r
        if (largest != i):
            self.swap(largest, i)
            self.max_heapify(largest)
 
    def swap(self, i, j):
        self.items[i], self.items[j] = self.items[j], self.items[i]
 
    def insert(self, key):
        index = self.size()
        self.items.append(key)
 
        while (index != 0):
            p = self.parent(index)
            if self.get(p) < self.get(index):
                self.swap(p, index)
            index = p
    
    # Min heap
    def MinInsert(self,key):
        index = self.size()
        self.items.append(key)
        
        while index!=0:
            p = self.parent(index)
            if self.get(p)> self.get(index):
                self.swap(p,index)
            index = p    
```
